```python
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - LAYOUT - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #

website = JupyterDash(__name__,external_stylesheets=[dbc.themes.SOLAR],
                    meta_tags=[{"name":"viewport","content":"width=device-width, initial-scale=1.0"}],
                   suppress_callback_exceptions=True)

SIDEBAR_STYLE = {
    "position": "fixed",
    "top": 0,
    "left": 0,
    "bottom": 0,
    "width": "16rem",
    "padding": "2rem 1rem",
}

CONTENT_STYLE = {
    "margin-left": "18rem",
    "margin-right": "2rem",
    "padding": "2rem 1rem",
}

content = html.Div(id="page-content", children=[], style=CONTENT_STYLE)

sidebar = html.Div(
    [
        html.H2("Number recognition program", className="display-4",style={'textAlign':'center'}),
        html.Hr(),
        html.P(
            "EFMat final project", className="side"
        ),
        dbc.Nav(
            [
                dbc.NavLink("Description", href="/", active="exact"),
                dbc.NavLink("Statistics", href="/page-1", active="exact"),
                dbc.NavLink("Canvas board", href="/page-2", active="exact"),
            ],
            vertical=True,
            pills=True,
        ),
    ],
    style=SIDEBAR_STYLE,

)


website.layout=html.Div([
    dbc.Button("Success",color="success", className="mr-1"),
    dbc.Row(dbc.Col(html.Title("Number recognition project")
    ),),
    dcc.Location(id="url"),
    sidebar,
    content    
    ])


# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - NRP - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - # 


df=pd.read_excel(r"C:\Users\adm\Documents\Jupyter\Semestre2\Programação\assets\numbers_base.xlsx")
df=df[["numb","height","width","x_co","y_co"]]


# ----------------------------------------------------------------------------------------------------------------------- #
# ---------------------------------------------------- DRAW ------------------------------------------------------------- #
# ----------------------------------------------------------------------------------------------------------------------- #

# ---------------------------------------------- DRAW NUM FILTER ---------------------------------------------- #



def NTP_draw(lstx,lsty,sizex,sizey):

    fig={"coordx_dist":{},"coordy_dist":{},
           "final_distW":{},"final_distR":{},"final_dist_abs":{},
           "width":{},"height":{},"length":{},
           "ratio_wh":{},"ratio_xy":{}, "ratio_x":{},"ratio_y":{},
           "StanDevx":{},"SStanDevx":{},
           "StanDevy":{},"SStanDevy":{},
           "method":{}}

    i=int(len(lstx)/10)

    basex=lstx[0]
    basey=lsty[0]

    final_distW=[]
    final_distRx=[]
    final_distRy=[]
    ratiod_x=[]
    ratiod_y=[]
    ratio_xy=[]
    stddx=[]
    stddy=[]

    for j in range(1,11):

        fig["coordx_dist"].update({

                "coordx"+str(j)+"_"+str(j+1):lstx[(j-1)*i:j*i]

                                })

        fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]=np.array(fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)])

        fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]=fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]-basex

        fig["coordy_dist"].update({

                "coordy"+str(j)+"_"+str(j+1):lsty[(j-1)*i:j*i]

                                })

        fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]=np.array(fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)])

        fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]=fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]-basey

        stddx.append(np.std(fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]))
        stddy.append(np.std(fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]))

        final_distRx.append(sum(fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]))
        ratiod_x.append(fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)])

        final_distRy.append(sum(fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]))
        ratiod_y.append(fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)])

    final_distRx=np.array(final_distRx)
    final_distRy=np.array(final_distRy)
    ratiod_x=np.array(ratiod_x)
    ratiod_y=np.array(ratiod_y)

    fig["final_distR"].update({

            "num":np.sqrt((sum(final_distRx))**2+(sum(final_distRy))**2) 

        })

    fig["ratio_xy"].update({

        "num":sum(final_distRx)/sum(final_distRy)

    })

    fig["ratio_x"].update({

        "num": ratiod_x

    })

    fig["ratio_y"].update({

        "num": ratiod_y

    })

    fig["ratio_wh"].update({

        "num": sizex/sizey

    })

    fig["length"].update({

        "num": len(lsty)

        })

    fig["StanDevx"].update({

        "num": stddx

        })

    fig["SStanDevx"].update({

        "num": np.std(stddx)

        })

    fig["StanDevy"].update({

        "num": stddy

        })

    fig["SStanDevy"].update({

        "num": np.std(stddy)

        })


    data2={#"Wrong":b_fig["final_distW"],
      "Right":fig["final_distR"],"Ratio_xy":fig["ratio_xy"],
      #"Abs":b_fig["final_dist_abs"],
      "Width":fig["width"],"Height":fig["height"],
      "Length":fig["length"],"Ratio_wh":fig["ratio_wh"],
      "StanDevx":fig["StanDevx"], "SStanDevx":fig["SStanDevx"],
      "StanDevy":fig["StanDevy"], "SStanDevy":fig["SStanDevy"],
      "Method":fig["method"]}

    dfff=pd.DataFrame.from_dict(data2)
    display(dfff)
    return fig, dfff, stddx, stddy


def NRP_base(fig,sizex,sizey,df,dfff,stddx,stddy,inf_limit,upper_limit):
    
    b_fig={"coordx_dist":{},"coordy_dist":{},
           "final_distW":{},"final_distR":{},"final_dist_abs":{},
           "width":{},"height":{},"length":{},
           "ratio_x":{},"ratio_y":{},"ratio_wh":{},"ratio_xy":{},
           "StanDevx":{},"SStanDevx":{},
           "StanDevy":{},"SStanDevy":{},
           "method":{}}
    
    for l in range(0,10):

        lstxx=df["x_co"][l]
        lstxx=lstxx.split(",")
        fx_lst=[float(item)for item in lstxx]

        lstyy=df["y_co"][l]
        lstyy=lstyy.split(",")
        fy_lst=[float(item)for item in lstyy]

        i=int(len(fy_lst)/10)

        b_fig["ratio_wh"].update({

                "num"+str(l): df["width"][l]/df["height"][l]

            })

        b_fig["length"].update({

                "num"+str(l): len(fy_lst)

            })

        basexx=fx_lst[0]
        baseyy=fy_lst[0]

        final_distW=[]
        final_distRx=[]
        final_distRy=[]
        ratiob_x=[]
        ratiob_y=[]
        ratio_xy=[]
        stdbx=[]
        stdby=[]


        for j in range(1,11):

     # ---------------------------------------------- X AXIS ---------------------------------------------- #

            b_fig["coordx_dist"].update({

                    "coordx"+str(j)+"_"+str(j+1):fx_lst[(j-1)*i:j*i]

                                    })

            b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]=np.array(b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)])

            b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]=b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]-basexx

            #---------------------------------------------------HERE
            #----------------------------------------------------|
            #----------------------------------------------------|
            #---------------------------------------------------\/
            b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]= sizex/df["width"][l] * b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]

     # ---------------------------------------------- Y AXIS ---------------------------------------------- #

            b_fig["coordy_dist"].update({

                    "coordy"+str(j)+"_"+str(j+1):fy_lst[(j-1)*i:j*i]

                                    })

            b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]=np.array(b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)])

            b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]=b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]-baseyy

            #---------------------------------------------------HERE
            #----------------------------------------------------|
            #----------------------------------------------------|
            #---------------------------------------------------\/
            b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]= sizey/df["height"][l]*b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]

     # ------------------------------------------- ANALYSIS ---------------------------------------------- #

            stdbx.append(np.std(b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]))
            stdby.append(np.std(b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]))

            final_distRx.append(sum(b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]))
            ratiob_x.append(b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)])

            final_distRy.append(sum(b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]))
            ratiob_y.append(b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)])

        final_distRx=np.array(final_distRx)    
        final_distRy=np.array(final_distRy) 
        ratiob_x=np.array(ratiob_x)
        ratiob_y=np.array(ratiob_y)


        b_fig["final_distR"].update({

            "num"+str(l):np.sqrt(sum(final_distRx)**2+sum(final_distRy)**2)

        })

        b_fig["ratio_xy"].update({

            "num"+str(l): sum(final_distRx) / sum(final_distRy)

        })

        b_fig["ratio_x"].update({

            "num"+str(l): ratiob_x

        })

        b_fig["ratio_y"].update({

            "num"+str(l): ratiob_y

        })

        b_fig["StanDevx"].update({

            "num"+str(l): stdbx

        })

        b_fig["SStanDevx"].update({

            "num"+str(l): np.std(stdbx)

        })

        b_fig["StanDevy"].update({

        "num"+str(l): stdby

        })

        b_fig["SStanDevy"].update({

            "num"+str(l): np.std(stdby)

        })


    data={#"Wrong":b_fig["final_distW"],
          "Right":b_fig["final_distR"],"Ratio_xy":b_fig["ratio_xy"],
          #"Abs":b_fig["final_dist_abs"],
          #"Width":b_fig["width"],"Height":b_fig["height"],
          "Length":b_fig["length"],"Ratio_wh":b_fig["ratio_wh"],
          "ratio_x":b_fig["ratio_x"], "ratio_y":b_fig["ratio_y"],
          "StanDevx":b_fig["StanDevx"], "SStanDevx":b_fig["SStanDevx"],
          "StanDevy":b_fig["StanDevy"], "SStanDevy":b_fig["SStanDevy"],
          #"Method":b_fig["method"]
         }

    dff=pd.DataFrame.from_dict(data)
    display(dff)


    # ------------------------------------------- STAN_DEV ANALYSIS ---------------------------------------------- #


    Analysis_Dx=[]
    Analysis_Dy=[]

    final_res_standevx=[]
    final_res_standevy=[]

    stddx=np.array(stddx)
    stddy=np.array(stddy)


    for i in range(0,10):
        result=[]
        divx=stddx/np.array(dff.iloc[i]["StanDevx"])


        if fig["ratio_xy"].get("num")>0 and dff.iloc[i]["Ratio_xy"]>0:
            for item in divx:
                if item >= inf_limit and item<=upper_limit:
                    result.append(1)

                else:
                    result.append(0)
            Analysis_Dx.append(sum(result))


        elif fig["ratio_xy"].get("num")>0 and dff.iloc[i]["Ratio_xy"]<0:
            Analysis_Dx.append(0)

        elif fig["ratio_xy"].get("num")<0 and dff.iloc[i]["Ratio_xy"]>0:
            Analysis_Dx.append(0)

        else:
            for item in divx:
                if item >= inf_limit and item<=upper_limit:
                    result.append(1)

                else:
                    result.append(0)

            Analysis_Dx.append(sum(result))


    for i in range(0,10):
        result=[]
        divy=stddy/np.array(dff.iloc[i]["StanDevy"])


        if fig["ratio_xy"].get("num")>0 and dff.iloc[i]["Ratio_xy"]>0:
            for item in divy:
                if item >= inf_limit and item<=upper_limit:
                    result.append(1)

                else:
                    result.append(0)
            Analysis_Dy.append(sum(result))


        elif fig["ratio_xy"].get("num")>0 and dff.iloc[i]["Ratio_xy"]<0:
            Analysis_Dy.append(0)

        elif fig["ratio_xy"].get("num")<0 and dff.iloc[i]["Ratio_xy"]>0:
            Analysis_Dy.append(0)

        else:
            for item in divy:
                if item >= inf_limit and item<=upper_limit:
                    result.append(1)

                else:
                    result.append(0)

            Analysis_Dy.append(sum(result))        


    nums=[e for e in range(10)]    

    dataFd={"nums":nums, "percentagex":Analysis_Dx,"percentagey":Analysis_Dy}

    df_d=pd.DataFrame.from_dict(dataFd)
    display(df_d)

    # ------------------------------------------- STAN_DEV RES ANALYSIS ---------------------------------------------- #


    if max(Analysis_Dx)<=2:
        final_res_standevx.append(Analysis_Dx.index(max(Analysis_Dx)))

    elif sum(df_d["percentagex"].nlargest(3))>=24:
        pos_r=df_d["percentagex"].nlargest(3)

        for p in range(len(df_d)):
            if df_d.iloc[p]["percentagex"]in pos_r.values:
                #print("+24-->",df_d.iloc[p])
                final_res_standevx.append(p)

            else:
                pass

    elif sum(df_d["percentagex"].nlargest(3))<24:
        pos_r=df_d["percentagex"].nlargest(3)

        if pos_r.iloc[-1]<=2:
            pos_r.drop(pos_r.index[2])

        if pos_r.iloc[1]<=2:
            pos_r.drop(pos_r.index[1])

        for p in range(len(df_d)):
            if df_d.iloc[p]["percentagex"]in pos_r.values:
                final_res_standevx.append(p)

            else:
                pass

    if max(Analysis_Dy)<=2:
        final_res_standevy.append(Analysis_Dy.index(max(Analysis_Dy)))

    elif sum(df_d["percentagey"].nlargest(3))>=24:
        pos_r=df_d["percentagey"].nlargest(3)

        for p in range(len(df_d)):
            if df_d.iloc[p]["percentagey"]in pos_r.values:
                final_res_standevy.append(p)

            else:
                pass

    elif sum(df_d["percentagey"].nlargest(3))<24:
        pos_r=df_d["percentagey"].nlargest(3)

        if pos_r.iloc[-1]<=2:
            pos_r.drop(pos_r.index[2])

        if pos_r.iloc[1]<=2:
            pos_r.drop(pos_r.index[1])

        for p in range(len(df_d)):
            if df_d.iloc[p]["percentagey"]in pos_r.values:
                final_res_standevy.append(p)

            else:
                pass

    # ------------------------------------------- RATIO ANALYSIS ---------------------------------------------- #   


    Analysis_R=[]

    final_res_ratio=[]


    for i in range(10):

        result_p=[]

        div=fig["ratio_xy"].get("num")/dff.iloc[i]["Ratio_xy"]

        if fig["ratio_xy"].get("num")>0 and dff.iloc[i]["Ratio_xy"]>0:

            #----------HERE------------------------------------------------HERE
            #-----------|---------------------------------------------------|
            #---------- |---------------------------------------------------|
            #----------\/--------------------------------------------------\/
            for l in range(10):
                
                divb=(sizex/df["width"][l]*sum(dff.iloc[i]["ratio_x"][l]))/(sizey/df["height"][l]*sum(dff.iloc[i]["ratio_y"][l]))
                divd=sum(fig["ratio_x"].get("num")[l])/sum(fig["ratio_y"].get("num")[l])
                div=divd/divb
                
                """print("Final division")
                print(div)"""

                """print("ratioB")
                print(divb)
                print("ratioD")
                print(divd)
                print("part -->"+str(i))
                print(div)"""

                if div >= inf_limit and div<=upper_limit:
                    result_p.append(1)

                else:
                    result_p.append(0)

            Analysis_R.append(sum(result_p))


        elif fig["ratio_xy"].get("num")>0 and dff.iloc[i]["Ratio_xy"]<0:
            Analysis_R.append(0)

        elif fig["ratio_xy"].get("num")<0 and dff.iloc[i]["Ratio_xy"]>0:
            Analysis_R.append(0)

        else:
            for l in range(10):

            #----------HERE------------------------------------------------HERE
            #-----------|---------------------------------------------------|
            #---------- |---------------------------------------------------|
            #----------\/--------------------------------------------------\/   
                divb=(sizex/df["width"][l]*sum(dff.iloc[i]["ratio_x"][l]))/(sizey/df["height"][l]*sum(dff.iloc[i]["ratio_y"][l]))
                divd=sum(fig["ratio_x"].get("num")[l])/sum(fig["ratio_y"].get("num")[l])
                div=divd/divb

                if div >= inf_limit and div <=upper_limit:
                    result_p.append(1)

                else:
                    result_p.append(0)

            Analysis_R.append(sum(result_p))

    dataFr={"nums":nums, "percentage":Analysis_R}

    df_r=pd.DataFrame.from_dict(dataFr)

    display(df_r)

    # ------------------------------------------- RATIO_RES ANALYSIS ---------------------------------------------- #   

    if max(Analysis_R)<=2:
        final_res_ratio.append(Analysis_R.index(max(Analysis_R)))

    elif sum(df_r["percentage"].nlargest(3))>=24:
        pos_r=df_r["percentage"].nlargest(3)

        for p in range(df_r):
            if df_r.iloc[p]["percentage"]in pos_r.values:
                final_res_ratio.append(p)

            else:
                pass

    elif sum(df_r["percentage"].nlargest(3))<24:
        pos_r=df_r["percentage"].nlargest(3)

        if pos_r.iloc[-1]<=2:
            pos_r.drop(pos_r.index[2])

        if pos_r.iloc[1]<=2:
            pos_r.drop(pos_r.index[2])

        for p in range(len(df_r)):
            if df_r.iloc[p]["percentage"]in pos_r.values:
                final_res_ratio.append(p)

            else:
                pass   

    # ------------------------------------------- FINAL RES ---------------------------------------------- #   

    partial_res=[]

    for item in final_res_standevx:
        if item in final_res_ratio:
            partial_res.append(item)
        else:
            pass

    for item in final_res_standevy:
        if item in partial_res:
            pass
        else:
            partial_res.append(item)
            pass

    print("Partial result" +str(partial_res))

    if len(partial_res)==1:
        print("Final result len 1--> "+str(partial_res[0]))


    elif len(partial_res)>1:
        answer=[(Analysis_R[partial_res[l]]+Analysis_Dx[partial_res[l]]+Analysis_Dy[partial_res[l]])/2 for l in range(len(partial_res))]
        if answer.count(max(answer))>1:
            inf_limit+=0.2
            upper_limit-=0.2
            NRP_base(fig,sizex,sizey,df,dfff,stddx,stddy,inf_limit,upper_limit)
            
        else:
            print("Final result med --> "+str(partial_res[answer.index(max(answer))]))
            return partial_res[answer.index(max(answer))]




    else:
        lista=[max(Analysis_R),max(Analysis_Dx),max(Analysis_Dy)]
        if lista.index(max(lista)) == 0:
            print("Final result--> "+str(max(final_res_ratio)))
            return max(final_res_ratio),max(Analysis_R)

        elif lista.index(max(lista)) == 1:
            print("Final result--> "+str(max(final_res_standevx)))
            return max(final_res_standevx),max(Analysis_Dx)

        else:
            print("Final result--> "+str(max(final_res_standevy)))
            return max(final_res_standevy),max(Analysis_Dy)


# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - CARDS - - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #


card1=dbc.Container([
    dbc.CardHeader(   
    dbc.Button("Objective", id="objBtt",style={"color":"goldenrod"},className="text-dark btn-block align-middle w-100")
    ),
    dbc.Collapse(
    dbc.Card([
    dbc.CardBody([
        html.P("""This is the final project of UnB second semester EFMat course coordenated
                    by teacher Lucas Loureiro. Thus, subjects such as python Classes,
                    web development, and html, as well as the ones theached before,
                    were applied on this website. """,className="card-text text-justify")
        ],style={"color":"goldenrod"}),
    dbc.CardImg(src="https://lh3.googleusercontent.com/a-/AOh14GhKKmrBlYU4jgK88G7fFV3XLuyNgI4Le1Okj9Ve=s75-c",bottom=True)])
        ,id="objAns",is_open=True)])


card2=dbc.Container([
    dbc.CardHeader( 
    dbc.Button("Program explanation", id="pgmBtt",style={"color":"goldenrod"},className="text-dark btn-block align-middle w-100")),
    dbc.Collapse(
        dbc.Card([
    dbc.CardBody([
        html.P("""The general idea of the NRP (Number Recognition Program) is to possiblity the identification of a number drawn on a board. Hence, after the user drawn
                    a number on the board, the program will collect pixels that were painted and reshape them to a json list with the intention of building a coordinate system. From that,
                    a comparison of the distance, of the first pixel to the others, from predefined shapes of numbers will be done. Since this distances can be interpreted as vectos, an 
                    analysis of its tangent and a comparison between the length of each vector for each axe were done to set a partial result. Having finished this step, a division of the vectors in groups of ten is done to increase the
                    precision of the analysis, since it is possible to check more precisely the draw shape when comparing each part direction tendence. Moreover, to evaluate the value interval of the numbers, the precision of each pool should be amid 0.6 and 1.5 to be qualified as a number candidate.
                    Once the tangent is periodic and has positive and negative values,
                    some numbers can be pre excluded of being cadidates as an acount of the draw tangent value. Next, the Standard Deviation of each vector pool is calculated to compare its variation with the base numbers Standard Deviation vectors pools.
                    Having concluded this step, the 3 highest precision numbers (of each analysis for each axe) are selected to be then summed in order to find the one with greatest media.
                    If two or more numbers have the same media, an recursion is done to restrict in 20 points the inferior and upper limits of the Confidence Interval. Finally, after all this analysis, the NRP declares the final result.
                    \n\n
                    """,
               className="card-text text-justify")
        
    ],style={"color":"goldenrod"}),
        dbc.CardImg(src="https://th.bing.com/th/id/OIP.hmcZgKUKYsC_5rQB1SmY_AHaE7?pid=ImgDet&rs=1",
                    bottom=True)]),id="pgmAns",is_open=True)])


card3=dbc.Container([
    dbc.CardHeader(
    dbc.Button("Instructions",id="intBtt",style={"color":"goldenrod"},className="text-dark btn-block align-middle w-100")),
    dbc.Collapse(
    dbc.Card([
    dbc.CardBody([
        html.P("""First, click on 'Canvas board' in the sidebar. Second, drawn a single number number within the yellow image on the center of the page.
                    Finished the drawing, click on 'Analize' button below the board. If some mistake was done or an improvement on the drawing should 
                    be done, click on the Left Arrow button below the board. Otherwise, just follow the previous step.\n\n
                    Finally, after clicking on 'Analize' button, you should select the number drawn on the dropdown bar and click on "Submit". 
                    After it, the program will declare a final result and ask you to check the NRP answer. Then, click on 'Yesh' or 'No' button below the result. If you want to make another drawn, just click on the Left Arrow button to rerun the program.\n\n
                    You may like to check the performace of the program. For this reason, the "Statistics" button on the sidebar was created. There you will find the general 
                    precision of the program, individual precision for each number, and most frequent numbers.
                    """,className="card-text text-justify")        
    ],style={"color":"goldenrod"}),dbc.CardImg(src="https://cdn0.iconfinder.com/data/icons/logistic-lineal-color/512/checklist-512.png",
                                               bottom=True)]),
        id="intAns",is_open=True)])


card_me=dbc.Container([
    dbc.CardHeader(
    dbc.Button("About me", style={"color":"goldenrod"}, className="text-dark btn-lg align-middle w-100 ", id="btme_btt")),
    dbc.Collapse(
        dbc.Card([
            dbc.CardBody([
                html.P("""- Name: Pedro Augusto Giacomelli Fernandes """,className="card-text"),
                           html.P("""- Profession: Student (UnB electrical engineer)""",className="card-text"), 
                           html.P("""- Contact e-mail: pedro.giacomelli2@gmail.com """,className="card-text"),
                        ],style={"color":"goldenrod"}
            ),
            html.Img(src=r"https://drive.google.com/file/d/1LnbHNc3uGQNChVs4ieA3kUC4GUWYdoHF/view"),            
            dbc.Button(
                        "My GitHub",
                        id="git_hub", 
                        className="ml-auto text-dark",
                        href='https://github.com/PedroGiacomelli/TRabalhoFinal',
                        style={"color":"goldenrod"}
                        )            
        ]),
    id="btme_ans",is_open=False)
])

cards=dbc.Container([
                    dbc.Row([
                        dbc.Col(card1,width={"size":3,"offsett":0},xs=12,sm=12,md=12,lg=3,xl=3
                        ),                        

                        dbc.Col(card2,width={"size":4,"offsett":0},xs=12,sm=12,md=12,lg=4,xl=4
                        ),
                        
                        dbc.Col(card3,width={"size":4,"offsett":0},xs=12,sm=12,md=12,lg=4,xl=4
                        )
                    ],justify="around"),
    
                    dbc.Row([
                        dbc.Col(card_me,width={"size":4,"offsett":0},xs=12,sm=12,md=12,lg=4,xl=4
                        ),
                        
                    ],justify="around")
    
                ])


# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - GRAPHS - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #


df1=pd.read_csv(r"C:\Users\adm\Documents\Jupyter\Semestre2\TRabalhoFinal\assets\program_precision")
df2=pd.read_csv(r"C:\Users\adm\Documents\Jupyter\Semestre2\TRabalhoFinal\assets\individual_precision")
df3=pd.read_csv(r"C:\Users\adm\Documents\Jupyter\Semestre2\TRabalhoFinal\assets\number_frequence")

graph1=dbc.Container([
        dbc.Button(html.H1("General program precision", className="text-dark mb-4"),
                                className="badge-pill", style={"color":"goldenrod"}),
        dcc.Graph(id="idg1",figure={})
                    ])
                     
graph2=dbc.Container([
        dbc.Button(html.H1("Individual number precision", className="text-dark mb-4"),
                                className="badge-pill", style={"color":"goldenrod"}),
        dcc.Graph(id="idg2",figure={})
        
                    ])

graph3=dbc.Container([
        dbc.Button(html.H1("General number frequency", className="text-dark mb-4"),
                                className="badge-pill", style={"color":"goldenrod"}),
        dcc.Graph(id="idg3",figure={})        
                    ])

graphs=dbc.Container([        
    
        dbc.Row([
            dbc.Col([graph1],width={"size":4},xs=12,sm=12,md=12,lg=4,xl=4),
            dbc.Col([graph2],width={"size":4},xs=12,sm=12,md=12,lg=4,xl=4),
            dbc.Col([graph3],width={"size":4},xs=12,sm=12,md=12,lg=4,xl=4)
        ],justify="around")
])



# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - CANVAS - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #


file="assets\DrawingBackGround.PNG"

#columns=["width","height"]

canvas1=html.Div([
    dbc.Row([
        dbc.Col([
            DashCanvas(id='canvas_board',
               goButtonTitle ="Analize",
               width=150,height=400,
               hide_buttons=['line',"rectangle","pan",'zoom',"select","redo"],
               lineWidth=5,
               lineColor="#073642",
               filename=file
                      )
            #dash_table.DataTable(id="drawn_table",columns=[{"name":i,"id":i} for i in columns])
        ],width={"size":5,"offset":5})
        ]),
    
    dbc.Row([
        html.Hr(),
        dbc.Col([html.H5("Choose the number drawn:"),
        dcc.Dropdown(
            id="dpdn",
            multi=False,
            value=0,
            options=[{"label":x,"value":x}for x in range(0,10)]),
                dbc.Button("Submit",id="sbm_btt",style={"color":"goldenrod"},
                           className="text-dark btn-block")],width={"size":3,"offset":0}
                )
],justify="center"),  ])  

column=["number","precision"]

canvas2=dbc.Container([
    dbc.Row([
        
        dbc.Col([html.H5("NRP suggested as the final answer the number :", className="text-center")], width={"size":2}),
        dash_table.DataTable(id="num_table",style_cell={'textAlign': 'left'},columns=[{"name":i,"id":i} for i in column])
    ]),
    
    dbc.Row([
        html.Hr(),
        dbc.Col([html.H5("Is the program's answer correct?",className="text-center")],width={"size":2}),
        dbc.Col([dbc.Button("YES",id="y_btt",style={"color":"goldenrod"},className="text-dark btn-block")],width={"size":1}),
        dbc.Col([dbc.Button("NO",id="n_btt",style={"color":"goldenrod"},className="text-dark btn-block")],width={"size":1,"offset":0})
        
    ],justify="center")
])
    


# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - CALLBACKS - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #


# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - COORDENATION - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #


@website.callback(
    Output("page-content", "children"),
    [Input("url", "pathname")]
)

def page_content(pathname):
    if pathname == "/":
        return [
                html.H1('PROJECT DESCRIPTION',
                        style={'textAlign':'center',"color":"goldenrod"}, className="mb-4"),
                cards
                
                ]
    elif pathname == "/page-1":
        return [
                html.H1('Program statistics',
                        style={'textAlign':'center',"color":"goldenrod"},className="mb-4"),
                graphs
                
                ]
    elif pathname == "/page-2":
        return [

                html.H1('CANVAS BOARD',
                style={'textAlign':'center',"color":"goldenrod"},className="mb-4"),

                html.H5("""Drawn the number inside the canvas.
                    Make sure to drawn the number on the center and as big as possible.""",
                    style={'textAlign':'center'}), canvas1, canvas2        
                ]
    
    
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - CARDS COLAPSE - - - - - - - - - - - - - - - - - - - - - - - - - - #     
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #


@website.callback(Output("objAns","is_open"),
                  [Input("objBtt","n_clicks")],
                  [State("objAns","is_open")],
                 )
def toggle_collapse(n,is_open):
    if n:
        return not is_open
    return is_open


@website.callback(Output("pgmAns","is_open"),
                  [Input("pgmBtt","n_clicks")],
                  [State("pgmAns","is_open")],
                 )
def toggle_collapse(n,is_open):
    if n:
        return not is_open
    return is_open


@website.callback(Output("intAns","is_open"),
                  [Input("intBtt","n_clicks")],
                  [State("intAns","is_open")],
                 )
def toggle_collapse(n,is_open):
    if n:
        return not is_open
    return is_open

@website.callback(Output("btme_ans","is_open"),
                  [Input("btme_btt","n_clicks")],
                  [State("btme_ans","is_open")],
                 )
def toggle_collapse(n,is_open):
    if n:
        return not is_open
    return is_open

# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - GRAPHS - - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #


"""
@website.callback(Output("idg3","children"),
                 [Intput("result","n_clicks"),Input("number","text")])

def update_graph(clicks):
    if clicks:
        df"""



# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - DRAWING DATA - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #

@website.callback(Output("num_table","data"),Input("canvas_board", "json_data"))

def drawn_data(string):
    
    if string:
        data=json.loads(string)

        lstx=[]
        lsty=[]
        
        for i in range(1,len(data["objects"][1]["path"])-1):
            
            lstx.append(data["objects"][1]["path"][i][1])
            lstx.append(data["objects"][1]["path"][i][3])
            lsty.append(data["objects"][1]["path"][i][2])
            lsty.append(data["objects"][1]["path"][i][4])

        sizex=data["objects"][1]["width"]
        sizey=data["objects"][1]["height"]
        NTP_draw(lstx,lsty,sizex,sizey)
        fig, dfff, stddx, stddy = NTP_draw(lstx,lsty,sizex,sizey)
        inf_limit=0.6
        upper_limit=1.5
        data=NRP_base(fig,sizex,sizey,df,dfff,stddx,stddy,inf_limit,upper_limit)
        return data


    else:
        raise PreventUpdate
    #return data['objects'][1:]


# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - CANVAS RES - - - - - - - - - - - - - - - - - - - - - - - - - - #
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #


@website.callback(
    Output("idg1","figure"),
    [Input("sbm_btt","n_clicks")],
    [State("dpdn","value"), State("num_table","data")],
)

def trig1 (n1,n2,n3):
    print(n2,n3)
    
    if n1:
        if int(n2) == int(n3):
            df1.loc[0,"Final"]= df1.loc[0,"Final"]+1
            df1[0,"Precision"]=df1[0,"Final"]/sum(df1["Final"])
            df1[1,"Precision"]=df1[1,"Final"]/sum(df1["Final"])
            bar1=px.bar(data_frame=df1,
               x="Result",y="Precision",
                orientation="v", color_discrete_sequence=["goldenrod"],
               )
            bar1.update_layout(paper_bgcolor="#002b36",plot_bgcolor="#002b36")

            return bar1
        
        else:
            df1.loc[1,"Final"]= df1.loc[1,"Final"]+1
            df1[0,"Precision"]=df1[0,"Final"]/sum(df1["Final"])
            df1[1,"Precision"]=df1[1,"Final"]/sum(df1["Final"])
            bar1=px.bar(data_frame=df1,
               x="Result",y="Precision",
                orientation="v", color_discrete_sequence=["goldenrod"],
               )
            bar1.update_layout(paper_bgcolor="#002b36",plot_bgcolor="#002b36")

            return bar1
    else:
        pass
    
    """if triggered_id=="sbm_btt":
        df1.loc[0,"Final"]= df1.loc[0,"Final"]+1
        df1[0,"Precision"]=df1[0,"Final"]/sum(df1["Final"])
        df1[1,"Precision"]=df1[1,"Final"]/sum(df1["Final"])
        bar1=px.bar(data_frame=df1,
           x="Result",y="Precision",
            orientation="v", color_discrete_sequence=["goldenrod"],
           )
        bar1.update_layout(paper_bgcolor="#002b36",plot_bgcolor="#002b36")
        
        return bar1
    """
    
    """ elif triggered_id=="n_btt":
        df1.loc[1,"Final"]= df1.loc[1,"Final"]+1
        df1[0,"Precision"]=df1[0,"Final"]/sum(df1["Final"])
        df1[1,"Precision"]=df1[1,"Final"]/sum(df1["Final"])
        bar1=px.bar(data_frame=df1,
           x="Result",y="Precision",
            orientation="v", color_discrete_sequence=["goldenrod"],
           )
        bar1.update_layout(paper_bgcolor="#002b36",plot_bgcolor="#002b36")
        return bar1
    
    
    else:
        df1[0,"Precision"]=df1[0,"Final"]/sum(df1["Final"])
        df1[1,"Precision"]=df1[1,"Final"]/sum(df1["Final"])
        bar1=px.bar(data_frame=df1,
           x="Result",y="Precision",
            orientation="v", color_discrete_sequence=["goldenrod"],
           )
        bar1.update_layout(paper_bgcolor="#002b36",plot_bgcolor="#002b36")
        
        return bar1"""

    
@website.callback(
    Output("idg2","figure"),
    [Input("y_btt","n_clicks"),Input("n_btt","n_clicks")],
    [State("dpdn","value")],
)
def trig2 (yb,nb,value):

    button_clicked = ctx.triggered_id
    print(button_clicked)
    
    if not button_clicked:
        
        df2.loc[value,"Precision"]=df2.loc[value,"Right"]/df2.loc[value,"Total"]
        
        bar2=px.bar(data_frame=df2,
           x="Numbers",y="Precision",
            orientation="v", color_discrete_sequence=["goldenrod"], labels={"Precision":"Precision in %"}
           )
        bar2.update_layout(paper_bgcolor="#002b36",plot_bgcolor="#002b36")

        
        return bar2
    
    elif button_clicked=="y_btt":
        
        df2.loc[value,"Total"]=df2.loc[value,"Total"]+1
        df2.loc[value,"Right"]=df2.loc[value,"Right"]+1
        df2.loc[value,"Precision"]=df2.loc[value,"Right"]/df2.loc[value,"Total"]
        
        bar2=px.bar(data_frame=df2,
           x="Numbers",y="Precision",
            orientation="v", color_discrete_sequence=["goldenrod"], labels={"Precision":"Precision in %"}
           )
        bar2.update_layout(paper_bgcolor="#002b36",plot_bgcolor="#002b36")

        return bar2
    
    elif button_clicked=="n_btt":
        
        df2.loc[value,"Total"]=df2.loc[value,"Total"]+1
        df2.loc[value,"Precision"]=df2.loc[value,"Right"]/df2.loc[value,"Total"]
        
        bar2=px.bar(data_frame=df2,
           x="Numbers",y="Precision",
            orientation="v", color_discrete_sequence=["goldenrod"], labels={"Precision":"Precision in %"}
           )
        bar2.update_layout(paper_bgcolor="#002b36",plot_bgcolor="#002b36")

        return bar2
    
@website.callback(
    Output("idg3","figure"),
    [Input("sbm_btt","n_clicks")],
    [State("dpdn","value")],
)
def trig3 (n,value):
    df3.loc[value,"Frequence"]= df3.loc[value,"Frequence"]+1
    bar3=px.bar(data_frame=df3,
       x="Numbers",y="Frequence",
        orientation="v", color_discrete_sequence=["goldenrod"], labels={"Frequence":"Frequency in %"}
       )
    bar3.update_layout(paper_bgcolor="#002b36",plot_bgcolor="#002b36")

    return bar3

if __name__ == '__main__':
    website.run_server(debug=True)
```

```python
# ----------------------------------------------------------------------------------------------------------------------- #
# ---------------------------------------------------- DRAW ------------------------------------------------------------- #
# ----------------------------------------------------------------------------------------------------------------------- #

# ---------------------------------------------- DRAW NUM FILTER ---------------------------------------------- #

def NTP_draw(lstx,lsty,preguiçax,preguiçay):

    fig={"coordx_dist":{},"coordy_dist":{},
           "final_distW":{},"final_distR":{},"final_dist_abs":{},
           "width":{},"height":{},"length":{},
           "ratio_wh":{},"ratio_xy":{}, "ratio_x":{},"ratio_y":{},
           "StanDevx":{},"SStanDevx":{},
           "StanDevy":{},"SStanDevy":{},
           "method":{}}

    i=int(len(lstx)/10)

    basex=lstx[0]
    basey=lsty[0]

    final_distW=[]
    final_distRx=[]
    final_distRy=[]
    ratiod_x=[]
    ratiod_y=[]
    ratio_xy=[]
    stddx=[]
    stddy=[]

    for j in range(1,11):

        fig["coordx_dist"].update({

                "coordx"+str(j)+"_"+str(j+1):lstx[(j-1)*i:j*i]

                                })

        fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]=np.array(fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)])

        fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]=fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]-basex

        fig["coordy_dist"].update({

                "coordy"+str(j)+"_"+str(j+1):lsty[(j-1)*i:j*i]

                                })

        fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]=np.array(fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)])

        fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]=fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]-basey

        stddx.append(np.std(fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]))
        stddy.append(np.std(fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]))

        final_distRx.append(sum(fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]))
        ratiod_x.append(fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)])

        final_distRy.append(sum(fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]))
        ratiod_y.append(fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)])

    final_distRx=np.array(final_distRx)
    final_distRy=np.array(final_distRy)
    ratiod_x=np.array(ratiod_x)
    ratiod_y=np.array(ratiod_y)

    fig["final_distR"].update({

            "num":np.sqrt((sum(final_distRx))**2+(sum(final_distRy))**2) 

        })

    fig["ratio_xy"].update({

        "num":sum(final_distRx)/sum(final_distRy)

    })

    fig["ratio_x"].update({

        "num": ratiod_x

    })

    fig["ratio_y"].update({

        "num": ratiod_y

    })

    fig["ratio_wh"].update({

        "num": preguiçax/preguiçay

    })

    fig["length"].update({

        "num": len(lsty)

        })

    fig["StanDevx"].update({

        "num": stddx

        })

    fig["SStanDevx"].update({

        "num": np.std(stddx)

        })

    fig["StanDevy"].update({

        "num": stddy

        })

    fig["SStanDevy"].update({

        "num": np.std(stddy)

        })


    data2={#"Wrong":b_fig["final_distW"],
      "Right":fig["final_distR"],"Ratio_xy":fig["ratio_xy"],
      #"Abs":b_fig["final_dist_abs"],
      "Width":fig["width"],"Height":fig["height"],
      "Length":fig["length"],"Ratio_wh":fig["ratio_wh"],
      "StanDevx":fig["StanDevx"], "SStanDevx":fig["SStanDevx"],
      "StanDevy":fig["StanDevy"], "SStanDevy":fig["SStanDevy"],
      "Method":fig["method"]}

    dfff=pd.DataFrame.from_dict(data2)
    display(dfff)
    return fig, dfff, stddx, stddy

NTP_draw(lstx,lsty,preguiçax,preguiçay)

fig=NTP_draw(lstx,lsty)[0]
dfff=NTP_draw(lstx,lsty)[1]
sttdx=NTP_draw(lstx,lsty)[2]
stddy=NTP_draw(lstx,lsty)[3]
```

```python
# ----------------------------------------------------------------------------------------------------------------------- #
# ---------------------------------------------------- BASE ------------------------------------------------------------- #
# ----------------------------------------------------------------------------------------------------------------------- #

# ---------------------------------------------- BASE NUMS FILTER ---------------------------------------------- #

def NRP_base(fig,preguiçax,preguiçay,df,dfff,stddx,stddy,inf_limit,upper_limit):
    
    b_fig={"coordx_dist":{},"coordy_dist":{},
           "final_distW":{},"final_distR":{},"final_dist_abs":{},
           "width":{},"height":{},"length":{},
           "ratio_x":{},"ratio_y":{},"ratio_wh":{},"ratio_xy":{},
           "StanDevx":{},"SStanDevx":{},
           "StanDevy":{},"SStanDevy":{},
           "method":{}}
    
    for l in range(0,10):

        lstxx=df["x_co"][l]
        lstxx=lstxx.split(",")
        fx_lst=[float(item)for item in lstxx]

        lstyy=df["y_co"][l]
        lstyy=lstyy.split(",")
        fy_lst=[float(item)for item in lstyy]

        i=int(len(fy_lst)/10)

        b_fig["ratio_wh"].update({

                "num"+str(l): df["width"][l]/df["height"][l]

            })

        b_fig["length"].update({

                "num"+str(l): len(fy_lst)

            })

        basexx=fx_lst[0]
        baseyy=fy_lst[0]

        final_distW=[]
        final_distRx=[]
        final_distRy=[]
        ratiob_x=[]
        ratiob_y=[]
        ratio_xy=[]
        stdbx=[]
        stdby=[]


        for j in range(1,11):

     # ---------------------------------------------- X AXIS ---------------------------------------------- #

            b_fig["coordx_dist"].update({

                    "coordx"+str(j)+"_"+str(j+1):fx_lst[(j-1)*i:j*i]

                                    })

            b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]=np.array(b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)])

            b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]=b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]-basexx

            #---------------------------------------------------HERE
            #----------------------------------------------------|
            #----------------------------------------------------|
            #---------------------------------------------------\/
            b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]= preguiçax/df["width"][l] * b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]

     # ---------------------------------------------- Y AXIS ---------------------------------------------- #

            b_fig["coordy_dist"].update({

                    "coordy"+str(j)+"_"+str(j+1):fy_lst[(j-1)*i:j*i]

                                    })

            b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]=np.array(b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)])

            b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]=b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]-baseyy

            #---------------------------------------------------HERE
            #----------------------------------------------------|
            #----------------------------------------------------|
            #---------------------------------------------------\/
            b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]= preguiçay/df["height"][l]*b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]

     # ------------------------------------------- ANALYSIS ---------------------------------------------- #

            stdbx.append(np.std(b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]))
            stdby.append(np.std(b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]))

            final_distRx.append(sum(b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)]))
            ratiob_x.append(b_fig["coordx_dist"]["coordx"+str(j)+"_"+str(j+1)])

            final_distRy.append(sum(b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)]))
            ratiob_y.append(b_fig["coordy_dist"]["coordy"+str(j)+"_"+str(j+1)])

        final_distRx=np.array(final_distRx)    
        final_distRy=np.array(final_distRy) 
        ratiob_x=np.array(ratiob_x)
        ratiob_y=np.array(ratiob_y)


        b_fig["final_distR"].update({

            "num"+str(l):np.sqrt(sum(final_distRx)**2+sum(final_distRy)**2)

        })

        b_fig["ratio_xy"].update({

            "num"+str(l): sum(final_distRx) / sum(final_distRy)

        })

        b_fig["ratio_x"].update({

            "num"+str(l): ratiob_x

        })

        b_fig["ratio_y"].update({

            "num"+str(l): ratiob_y

        })

        b_fig["StanDevx"].update({

            "num"+str(l): stdbx

        })

        b_fig["SStanDevx"].update({

            "num"+str(l): np.std(stdbx)

        })

        b_fig["StanDevy"].update({

        "num"+str(l): stdby

        })

        b_fig["SStanDevy"].update({

            "num"+str(l): np.std(stdby)

        })


    data={#"Wrong":b_fig["final_distW"],
          "Right":b_fig["final_distR"],"Ratio_xy":b_fig["ratio_xy"],
          #"Abs":b_fig["final_dist_abs"],
          #"Width":b_fig["width"],"Height":b_fig["height"],
          "Length":b_fig["length"],"Ratio_wh":b_fig["ratio_wh"],
          "ratio_x":b_fig["ratio_x"], "ratio_y":b_fig["ratio_y"],
          "StanDevx":b_fig["StanDevx"], "SStanDevx":b_fig["SStanDevx"],
          "StanDevy":b_fig["StanDevy"], "SStanDevy":b_fig["SStanDevy"],
          #"Method":b_fig["method"]
         }

    dff=pd.DataFrame.from_dict(data)
    display(dff)


    # ------------------------------------------- STAN_DEV ANALYSIS ---------------------------------------------- #


    Analysis_Dx=[]
    Analysis_Dy=[]

    final_res_standevx=[]
    final_res_standevy=[]

    stddx=np.array(stddx)
    stddy=np.array(stddy)

    """print("stanDx")
    print(stddx)

    print("stanDy")
    print(stddy)"""

    for i in range(0,10):
        result=[]
        divx=stddx/np.array(dff.iloc[i]["StanDevx"])

        """print("stanB")
        print(np.array(dff.iloc[i]["StanDevx"]))
        print("num -->"+str(i))
        print(divx)"""

        if fig["ratio_xy"].get("num")>0 and dff.iloc[i]["Ratio_xy"]>0:
            for item in divx:
                if item >= inf_limit and item<=upper_limit:
                    result.append(1)

                else:
                    result.append(0)
            Analysis_Dx.append(sum(result))


        elif fig["ratio_xy"].get("num")>0 and dff.iloc[i]["Ratio_xy"]<0:
            Analysis_Dx.append(0)

        elif fig["ratio_xy"].get("num")<0 and dff.iloc[i]["Ratio_xy"]>0:
            Analysis_Dx.append(0)

        else:
            for item in divx:
                if item >= inf_limit and item<=upper_limit:
                    result.append(1)

                else:
                    result.append(0)

            Analysis_Dx.append(sum(result))


    for i in range(0,10):
        result=[]
        divy=stddy/np.array(dff.iloc[i]["StanDevy"])

        """print("stanB")
        print(np.array(dff.iloc[i]["StanDevy"]))
        print("num -->"+str(i))
        print(divy)"""

        if fig["ratio_xy"].get("num")>0 and dff.iloc[i]["Ratio_xy"]>0:
            for item in divy:
                if item >= inf_limit and item<=upper_limit:
                    result.append(1)

                else:
                    result.append(0)
            Analysis_Dy.append(sum(result))


        elif fig["ratio_xy"].get("num")>0 and dff.iloc[i]["Ratio_xy"]<0:
            Analysis_Dy.append(0)

        elif fig["ratio_xy"].get("num")<0 and dff.iloc[i]["Ratio_xy"]>0:
            Analysis_Dy.append(0)

        else:
            for item in divy:
                if item >= inf_limit and item<=upper_limit:
                    result.append(1)

                else:
                    result.append(0)

            Analysis_Dy.append(sum(result))        


    nums=[e for e in range(10)]    

    dataFd={"nums":nums, "percentagex":Analysis_Dx,"percentagey":Analysis_Dy}

    df_d=pd.DataFrame.from_dict(dataFd)
    display(df_d)

    # ------------------------------------------- STAN_DEV RES ANALYSIS ---------------------------------------------- #


    if max(Analysis_Dx)<=2:
        final_res_standevx.append(Analysis_Dx.index(max(Analysis_Dx)))

    elif sum(df_d["percentagex"].nlargest(3))>=24:
        pos_r=df_d["percentagex"].nlargest(3)

        for p in range(len(df_d)):
            if df_d.iloc[p]["percentagex"]in pos_r.values:
                #print("+24-->",df_d.iloc[p])
                final_res_standevx.append(p)

            else:
                pass

    elif sum(df_d["percentagex"].nlargest(3))<24:
        pos_r=df_d["percentagex"].nlargest(3)

        if pos_r.iloc[-1]<=2:
            pos_r.drop(pos_r.index[2])

        if pos_r.iloc[1]<=2:
            pos_r.drop(pos_r.index[1])

        for p in range(len(df_d)):
            if df_d.iloc[p]["percentagex"]in pos_r.values:
                #print("-24-->",df_d.iloc[p])
                final_res_standevx.append(p)

            else:
                pass

    if max(Analysis_Dy)<=2:
        final_res_standevy.append(Analysis_Dy.index(max(Analysis_Dy)))

    elif sum(df_d["percentagey"].nlargest(3))>=24:
        pos_r=df_d["percentagey"].nlargest(3)

        for p in range(len(df_d)):
            if df_d.iloc[p]["percentagey"]in pos_r.values:
                #print("+24-->",df_d.iloc[p])
                final_res_standevy.append(p)

            else:
                pass

    elif sum(df_d["percentagey"].nlargest(3))<24:
        pos_r=df_d["percentagey"].nlargest(3)

        if pos_r.iloc[-1]<=2:
            pos_r.drop(pos_r.index[2])

        if pos_r.iloc[1]<=2:
            pos_r.drop(pos_r.index[1])

        for p in range(len(df_d)):
            if df_d.iloc[p]["percentagey"]in pos_r.values:
                #print("-24-->",df_d.iloc[p])
                final_res_standevy.append(p)

            else:
                pass

    # ------------------------------------------- RATIO ANALYSIS ---------------------------------------------- #   


    Analysis_R=[]

    final_res_ratio=[]


    for i in range(10):

        result_p=[]

        div=fig["ratio_xy"].get("num")/dff.iloc[i]["Ratio_xy"]

        #print(div)
        

        if fig["ratio_xy"].get("num")>0 and dff.iloc[i]["Ratio_xy"]>0:

            #----------HERE------------------------------------------------HERE
            #-----------|---------------------------------------------------|
            #---------- |---------------------------------------------------|
            #----------\/--------------------------------------------------\/
            for l in range(10):
                
                divb=(preguiçax/df["width"][l]*sum(dff.iloc[i]["ratio_x"][l]))/(preguiçay/df["height"][l]*sum(dff.iloc[i]["ratio_y"][l]))
                divd=sum(fig["ratio_x"].get("num")[l])/sum(fig["ratio_y"].get("num")[l])
                div=divd/divb
                
                """print("Final division")
                print(div)"""

                """print("ratioB")
                print(divb)
                print("ratioD")
                print(divd)
                print("part -->"+str(i))
                print(div)"""

                if div >= inf_limit and div<=upper_limit:
                    result_p.append(1)

                else:
                    result_p.append(0)

            Analysis_R.append(sum(result_p))


        elif fig["ratio_xy"].get("num")>0 and dff.iloc[i]["Ratio_xy"]<0:
            Analysis_R.append(0)

        elif fig["ratio_xy"].get("num")<0 and dff.iloc[i]["Ratio_xy"]>0:
            Analysis_R.append(0)

        else:
            for l in range(10):

            #----------HERE------------------------------------------------HERE
            #-----------|---------------------------------------------------|
            #---------- |---------------------------------------------------|
            #----------\/--------------------------------------------------\/   
                divb=(preguiçax/df["width"][l]*sum(dff.iloc[i]["ratio_x"][l]))/(preguiçay/df["height"][l]*sum(dff.iloc[i]["ratio_y"][l]))
                divd=sum(fig["ratio_x"].get("num")[l])/sum(fig["ratio_y"].get("num")[l])
                div=divd/divb
                """print("Final division num-->"+str(i))
                print(div)"""

                """print("ratioB")
                print(divb)
                print("ratioD")
                print(divd)
                print("part -->"+str(i))
                print(div)"""

                if div >= inf_limit and div <=upper_limit:
                    result_p.append(1)

                else:
                    result_p.append(0)

            Analysis_R.append(sum(result_p))

    dataFr={"nums":nums, "percentage":Analysis_R}

    df_r=pd.DataFrame.from_dict(dataFr)

    display(df_r)

    # ------------------------------------------- RATIO_RES ANALYSIS ---------------------------------------------- #   

    if max(Analysis_R)<=2:
        final_res_ratio.append(Analysis_R.index(max(Analysis_R)))

    elif sum(df_r["percentage"].nlargest(3))>=24:
        pos_r=df_r["percentage"].nlargest(3)

        for p in range(df_r):
            if df_r.iloc[p]["percentage"]in pos_r.values:
                final_res_ratio.append(p)

            else:
                pass

    elif sum(df_r["percentage"].nlargest(3))<24:
        pos_r=df_r["percentage"].nlargest(3)
        #print(pos_r)

        if pos_r.iloc[-1]<=2:
            pos_r.drop(pos_r.index[2])

        if pos_r.iloc[1]<=2:
            pos_r.drop(pos_r.index[2])

        for p in range(len(df_r)):
            if df_r.iloc[p]["percentage"]in pos_r.values:
                final_res_ratio.append(p)

            else:
                pass   

    # ------------------------------------------- FINAL RES ---------------------------------------------- #   

    partial_res=[]
    print("FinalStanDevx ",final_res_standevx)
    print("FinalStanDevy ",final_res_standevy)
    print("FinalRatio ",final_res_ratio)

    for item in final_res_standevx:
        if item in final_res_ratio:
            partial_res.append(item)
        else:
            pass

    for item in final_res_standevy:
        if item in partial_res:
            pass
        else:
            partial_res.append(item)
            pass

    print("Partial result" +str(partial_res))

    if len(partial_res)==1:
        print("Final result len 1--> "+str(partial_res[0]))


    #not the number
    elif len(partial_res)>1:
        answer=[(Analysis_R[partial_res[l]]+Analysis_Dx[partial_res[l]]+Analysis_Dy[partial_res[l]])/2 for l in range(len(partial_res))]
        #print("Answer"+str(answer))
        if answer.count(max(answer))>1:
            inf_limit+=0.2
            upper_limit-=0.2
            NRP_base(fig,preguiçax,preguiçay,df,dfff,stddx,stddy,inf_limit,upper_limit)
            
        else:
            print("Final result med --> "+str(partial_res[answer.index(max(answer))]))
            return partial_res[answer.index(max(answer))]




    else:
        lista=[max(Analysis_R),max(Analysis_Dx),max(Analysis_Dy)]
        if lista.index(max(lista)) == 0:
            print("Final result--> "+str(max(final_res_ratio)))
            return max(final_res_ratio)

        elif lista.index(max(lista)) == 1:
            print("Final result--> "+str(max(final_res_standevx)))
            return max(final_res_standevx)

        else:
            print("Final result--> "+str(max(final_res_standevy)))
            return max(final_res_standevy)

NRP_base(fig,preguiçax,preguiçay,df,dfff,stddx,stddy,inf_limit,upper_limit)
```
